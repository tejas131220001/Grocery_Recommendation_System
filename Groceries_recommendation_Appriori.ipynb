```python
df=pd.read_csv('Groceries_data.csv')
```

```python
df=df.sort_values(['Member_number','Date'])
```

```python
df.groupby(['Member_number','Date'])['itemDescription'].apply(lambda x:','.join(x)).reset_index()
```

```python
df['item_bought']=df.groupby(['Member_number','Date'])['itemDescription'].transform(lambda x:','.join(x))
```

```python
df
```

```python
df=df.drop('itemDescription',axis=1)
```

```python
df=df.drop_duplicates()
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec=CountVectorizer()
```

```python
temp_vec=vec.fit(df['item_bought'].apply(lambda x:' '.join( map(str,x.split(',')))))
temp_vec
```

```python
temp_vec.get_feature_names_out()
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
items=df['item_bought'].apply(lambda x:x.split(','))
items
```

```python
items_list=items.to_list()
items_list
```

```python
te=TransactionEncoder()
```

```python
itemset=pd.DataFrame(te.fit_transform(items_list),columns=te.columns_)
itemset
```

```python
from mlxtend.frequent_patterns import apriori
freq_itemset=apriori(itemset,min_support=0.001,use_colnames=True).sort_values('support',ascending=False)
freq_itemset['length']=freq_itemset['itemsets'].apply(lambda x:len(x))
```
